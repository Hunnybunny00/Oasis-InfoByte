```python
# Importing common machine learning models

from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.neural_network import MLPClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.ensemble import AdaBoostClassifier
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.mixture import GaussianMixture

from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
X = df1.drop(columns=['Species'])
Y = df1['Species']
x_train, x_test, y_train, y_test = train_test_split(X, Y,test_size = 42)
```

```python
x_train.shape
```

```python
y_test.shape
```

```python
y_train.shape
```

```python
x_test.shape
```

```python
logmodel = LogisticRegression()
logmodel.fit(x_train, y_train)
logreg_Accuracy = logmodel.score(x_test,y_test) * 100
print("Accuracy (Logistic Regression): ", logreg_Accuracy)
```

```python
dtcmodel = DecisionTreeClassifier()
dtcmodel.fit(x_train, y_train)

dtc_Accuracy = dtcmodel.score(x_test, y_test) * 100

print("Accuracy (DecisionTreeClassifier): ", dtc_Accuracy)
```

```python
rfcmodel = RandomForestClassifier()
rfcmodel.fit(x_train , y_train)
rfc_Accuracy = rfcmodel.score(x_test,y_test) * 100
print("Accuracy (RandomForestClassifier): ",rfc_Accuracy)
```

```python
svcmodel = SVC()
svcmodel.fit(x_train,y_train)
svc_pred = svcmodel.predict(x_test)
scv_Accuracy = svcmodel.score(x_test, y_test) * 100
print("Accuracy (SCV): ", scv_Accuracy)
classification_rep = classification_report(y_test, svc_pred)
classification_rep
```

```python
kncmodel = KNeighborsClassifier()
kncmodel.fit(x_train , y_train)
knc_Accuracy = kncmodel.score(x_test,y_test) * 100
kncmodel_pred = kncmodel.predict(x_test)
print("Accuracy (KNeightborsClassifier): ",knc_Accuracy)



# Evaluate the model
classification_rep = classification_report(y_test, kncmodel_pred)
classification_rep

```

```python
GNBmodel = GaussianNB()
GNBmodel.fit(x_train , y_train)
GNB_Accuracy = GNBmodel.score(x_test,y_test) * 100
print("Accuracy (GaussinNB): ",GNB_Accuracy)
```

```python
MLPCmodel = MLPClassifier()
MLPCmodel.fit(x_train , y_train)
MLPC_Accuracy = MLPCmodel.score(x_test,y_test) * 100
print("Accuracy (MLPClassifier): ",MLPC_Accuracy)
```

```python
GBCmodel = GradientBoostingClassifier()
GBCmodel.fit(x_train , y_train)
GBC_Accuracy = GBCmodel.score(x_test,y_test) * 100
print("Accuracy (GradientBoostingClassifier): ",GBC_Accuracy)
```

```python
ABCmodel = AdaBoostClassifier()
ABCmodel.fit(x_train , y_train)
ABC_Accuracy = ABCmodel.score(x_test,y_test) * 100
print("Accuracy (AdaBoosClassifire): ",ABC_Accuracy)
```

```python
best_accuracy_model = {
    'models': ["LogisticRegression",
         "DecisionTreeClassifier",
         "RandomForestClassifier",
         "SVC",
         "KNeighborsClassifier",
         "GaussianNB",
         "MLPClassifier",
         "GradientBoostingClassifier",
         "AdaBoostClassifier"],
    'accuracies': [logreg_Accuracy,
            dtc_Accuracy,
            rfc_Accuracy,
            scv_Accuracy,
            knc_Accuracy,
            GNB_Accuracy,
            MLPC_Accuracy,
            GBC_Accuracy,
            ABC_Accuracy],
    
}
```

```python
best_accuracy_model = pd.DataFrame(best_accuracy_model)
best_accuracy_model
```

```python
from tabulate import tabulate

print(tabulate(best_accuracy_model, headers='keys', tablefmt='pretty'))
```
